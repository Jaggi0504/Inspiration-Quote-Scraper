```python
import bs4
import requests
```

```python
url = "https://www.passiton.com/inspirational-quotes"
response = requests.get(url)
response.content
```

```python
soup = bs4.BeautifulSoup(response.content)
print(soup)
```

```python
row_element = soup.findAll('div', {'class': 'col-6'})
print(len(row_element))
```

```python
row = row_element[0]
row.img
```

```python
row.img.attrs['src']
```

```python
with open('inspiration.jpg', 'wb') as file:
    img_url = row.img.attrs['src']
    response = requests.get(img_url)
    file.write(response.content)
```

```python
for i, article in enumerate(row_element):
    with open('inspiration{}.jpg'.format(i), 'wb') as file:
        img_url = article.img.attrs['src']
        response = requests.get(img_url)
        file.write(response.content)
```
